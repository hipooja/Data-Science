```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Mall_Customers.csv')
df
```

```python
x = df.iloc[:,[3,4]]
x
```

```python
plt.scatter(x.iloc[:,0], x.iloc[:,1])
```

```python
# SSE -Sum Squared Error
```

```python
#elbow point 
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3, random_state=0)
```

```python
km.fit_predict(x)
```

```python
km.inertia_   #contain value of inertia (SSE)
```

```python
sse = []
for k in range(1,16):
    km =KMeans(n_clusters=k, random_state=0)
    km.fit_predict(x)
    sse.append(km.inertia_)
```

```python
plt.xlabel('Value of K')
plt.ylabel('SSE')
plt.grid()
plt.xticks(range(1,16))
plt.plot(range(1,16), sse, marker ='o', color='r')
```

```python
km = KMeans(n_clusters=5, random_state=0)
```

```python
y = km.fit_predict(x)
y
```

```python
plt.subplot(1,2,1)
plt.scatter(x.iloc[:,0], x.iloc[:,1])
plt.subplot(1,2,2)
plt.scatter(x.iloc[:,0], x.iloc[:,1], c=y)
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color='r')
```

```python
df['label']=y
sns.jointplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=df, hue='label', palette='tab10')	 
```
