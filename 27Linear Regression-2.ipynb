```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Salary_Data.csv')
df
```

```python
x = df[['YearsExperience']]
y = df['Salary']
```

```python
df.corr()
```

```python
plt.title('Experience  & Salary')
plt.grid()
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.scatter(x, y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(x, y)
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

```python
regressor.predict([[16]])
```

```python
regressor.predict(pd.DataFrame({'YearsExperience':[16]}))
```

```python
y_pred = regressor.predict(x)
```

```python
y_pred
```

```python
plt.title('Experience  & Salary')
plt.grid()
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.scatter(x, y, color='r', label='actual')
plt.plot(x, y_pred, color='g', label='predicated')
plt.legend()
```

```python
result = pd.DataFrame({
    'Actual': y,
    'Predicated': y_pred
})
result
```

```python
# mean absolute error
np.mean(abs(y - y_pred))
```

```python
# R-Square Error
regressor.score(x, y)
```
