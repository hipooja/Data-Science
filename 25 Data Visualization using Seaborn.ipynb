```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
iris = sns.load_dataset('iris')
```

```python
sns.displot(iris['sepal_length']);
```

```python
sns.kdeplot(iris['sepal_length']);
```

```python
sns.histplot(iris['sepal_length'], kde=True)
```

#### Joint Plot

```python
sns.jointplot(x = 'petal_length', y='petal_width', data=iris)
```

```python
sns.jointplot(x = 'petal_length', y='petal_width', data=iris, hue='species')
```

```python
import matplotlib.pyplot as plt
```

```python
help(sns.jointplot)
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue='species')
```

#### Strip Plot

```python
sns.stripplot(x = 'species', y = 'petal_length', data = iris, hue='species')
```

```python
sns.stripplot(x = 'species', y = 'petal_length', data = iris, hue='species', jitter=False)
```

#### Swarm plot

```python
sns.swarmplot(x = 'species', y = 'petal_length', data = iris, hue='species')
```

#### Box Plot

```python
sns.boxenplot(x = 'species', y = 'petal_length', data = iris, hue='species')
```
